```python
import os
import sys
from pathlib import Path

base_dir=f'{Path(os.getcwd()).parent}'
sys.path.append(f"{base_dir}")
```

```python
from Style_Bert_VITS2.common.tts_model import Model, ModelHolder
from Style_Bert_VITS2.config import config

def load_models(model_holder: ModelHolder):
    model_holder.models = []
    for model_name, model_paths in model_holder.model_files_dict.items():
        model = Model(
            model_path=model_paths[0],
            config_path=os.path.join(model_holder.root_dir, model_name, "config.json"),
            style_vec_path=os.path.join(
                model_holder.root_dir, model_name, "style_vectors.npy"
            ),
            device=model_holder.device,
        )
        model.load_net_g()
        model_holder.models.append(model)

model_holder = ModelHolder(f"{base_dir}/Style_Bert_VITS2/model_assets", "cuda:0")
load_models(model_holder)
print(model_holder.model_names)
```

```python
import pyopenjtalk

pyopenjtalk.run_frontend("テスト")
```

```python
model=model_holder.models[0]
print(f"style_id: {model.style2id.keys()}")
text="こんにちは。これは人工音声です"
sr, audio = model.infer(
    text=text,
    language="JP",
    sid=0,
    reference_audio_path=None,
    sdp_ratio=0.2,
    noise=0.6,
    noisew=0.8,
    length=1.0,
    line_split=True,
    split_interval=0.5,
    assist_text=None,
    assist_text_weight=1.0,
    use_assist_text=False,
    style="Fear",
    style_weight=15.0
)
import IPython.display as ipd
ipd.Audio(audio, rate=sr)
```
