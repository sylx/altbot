```python
import os
import sys
from pathlib import Path

base_dir=f'{Path(os.getcwd()).parent}'
sys.path.append(base_dir)

from typing import Tuple
import numpy as np

from libs.vits_japros.model import VITSJaProsModel
from libs.vits_japros.text import g2p
```

```python
model_name= "espnet"
model_path = f"{base_dir}/vits_models/{model_name}/192epoch.pth"
yaml_file = f"{base_dir}/vits_models/{model_name}/config.yaml"
device="gpu"
model = VITSJaProsModel(model_name, model_path, yaml_file, device=device)
```

```python
text="こんにちは、私は日本語の音声合成を行うシステムです。"
p=g2p(text)
print(p)
```

```python
speed_scale = 1.0
pitch_scale = 1.0
intonation_scale = 1.0
noise_scale = 0.0
noise_scale_dur = 0.0
sample_rate,output_audio=model.p2speech(
        p, speed_scale, pitch_scale, intonation_scale, noise_scale, noise_scale_dur
    )
```

```python
import IPython.display as ipd
ipd.Audio(output_audio, rate=sample_rate)
```
