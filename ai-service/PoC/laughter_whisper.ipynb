```python
import sys
import os
from pathlib import Path

base_dir=f'{Path(os.getcwd()).parent}'
sys.path.append(base_dir)

from faster_whisper import WhisperModel

model=WhisperModel(
    "base",
    device="cuda",
    compute_type="float32",
    download_root=f"{base_dir}/.model_cache",
)
```

```python
dir="/mnt/e/Download/laughterscape_ver1.0/ver1.0/denoised"
files=os.listdir(dir)
len(files)
```

```python
import librosa

prompt="笑"

for file in files[0:100]:
    # wavファイルを読み込む
    audio, sr = librosa.load(f"{dir}/{file}", sr=24000)
    # 16KHzにダウンサンプリング
    audio = librosa.resample(audio, orig_sr=sr, target_sr=16000)
    segments,info = model.transcribe(audio,
                                            condition_on_previous_text=False,
                                            prefix="",initial_prompt=prompt,
                                            compression_ratio_threshold=1.5,
                                            beam_size=3,language="ja",vad_filter=False,temperature=[0.0,0.2,0.4],best_of=2,
                                            word_timestamps=False)
    segments = list(segments) # generatorをlistに変換することでcoroutineを実行する
    if len(segments) == 0:
        print(f"{file}: No text")
        continue
    whole_text = ''.join([s.text for s in segments])
    temperature = [s.temperature for s in segments]
    compression_ratio = sum([s.compression_ratio for s in segments]) / len(segments)

    if compression_ratio > 1.5:
        print(f"{file}: decode error")
        continue

    print(f"{file}: {whole_text} {len(segments)} {temperature} {compression_ratio}")
```
